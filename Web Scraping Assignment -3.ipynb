```python
start_page = 0
end_page = 3
urls = []
for page in range(start_page,end_page):                                    #for scraping 3 pages
    try:
        page_urls = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-no-outline"]')         #scraping all the urls in the page for search result
        
        # appending all the urls on current page to urls list
        for i in page_urls:
            urls.append(i.get_attribute('href'))    # Scraping the url from webelement
            
        print("Product urls of page {} has been scraped.".format(page+1))                                  #printing scrraped pages
        
        # Moving to next page
        # Locating the next_button which is active
        nxt_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')      
        if nxt_button.text == 'Next':                                            # Checking if the button located is next button
            nxt_button.click()                                                    # Clicking the next button
            time.sleep(5)                                                         # time delay of 5 seconds
        # If the current active button is not next button, the we will check if the next button is inactive or not    
        elif driver.find_element(By.XPATH,'//span[@class="s-pagination-item s-pagination-next s-pagination-disabled "]').text == 'Next→':    
            print("No new pages exist. Breaking the loop")  # Printing message and breakinf loop if we have reached the last page
            break
            
    except NoSuchElementException as e:             # Handling NoSuchElementException    
        print("Stale Exception")
        next_page = nxt_button.get_attribute('href')        # Extracting the url of next page
        driver.get(next_page)                               # ReLoading the next page
```

# 10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ 


in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, 
overall reviews, privates from price, dorms from price, facilities and property description

```python
#importing the selenium webdriver
from selenium import webdriver

#Activating chrome browser
driver=webdriver.Chrome("Chromedriver.exe")

#lets Automate the searchbar
driver.maximize_window()
```

```python
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
```

```python
url="https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2022-11-21&to=2022-11-24&guests=2&page=1"
driver.get(url)                        #loading webpage
time.sleep(3)
```

```python
#creating web pages fpr storing scraped data
hostel_name=[]
distance=[]
ratings=[]
tot_reviews=[]
overall_reviews=[]
privates_price=[]
dorms_price=[]
description=[]
price=[]
```

```python

for i in driver.find_elements(By.XPATH,'//h2[@class="title title-6"]/a'):             #scarping all hostel names
    hostel_name.append(i.text)
    
for i in driver.find_elements(By.XPATH,'//span[@class="description"]'):             #scarping  distance from city
    distance.append(i.text)
    
for i in driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]'):             #scarping description
    description.append(i.text.replace('\n',"      "))
    
for i in driver.find_elements(By.XPATH,'//div[@class="rating rating-summary-container big"]'):      #scarping ratings
    ratings.append(i.text.split('\n')[0])
    
for i in driver.find_elements(By.XPATH,'//div[@class="reviews"] '):                  #scarping all total reviews
    tot_reviews.append(i.text)
    
for i in driver.find_elements(By.XPATH,'//div[@class="price-col"]'):                  #scarping  prices for both dorm and private
    price.append(i.text.replace('\n',"  ")) 
    
for i in driver.find_elements(By.XPATH,'//div[@class="keyword"]'):                  #scarping  overall reviews
    overall_reviews.append(i.text)    
   
```

```python
#skipping the privates price and storing dorms  price
dorms_price=price[1::2]
```

```python
#skipping the dorm price and storing privates price
privates_price=price[::2]
```

```python
#removing unwanted data
overall_reviews=overall_reviews[2:]
```
